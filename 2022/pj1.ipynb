### Thống kê mô tả tập dữ liệu

Tập dữ liệu Insurance là tập dữ liệu chi phí y tế cá nhân do bảo hiểm y tế thanh toán và các thông tin về bệnh nhân

- Age: Tuổi của bệnh nhân
- Sex: Giới tính (Nam / Nữ)
- BMI: Chỉ số khối cơ thể
- Children: Số trẻ em hưởng bảo hiểm y tế / Số người phụ thuộc
- Smoker: (Có / Không) hút thuốc
- Region: Khu vực sinh sống
- Charges: Chi phí y tế cá nhân do bảo hiểm y tế thanh toán

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Đọc dữ liệu

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/tranlybuu/BaiTap/main/2022/insurance.csv")
df1.head(5)
```

Mã hóa các biến của cột Sex - Smoker - Region

```python
from sklearn import preprocessing
df = df1[df1.columns[:]].apply(preprocessing.LabelEncoder().fit_transform)
```

Kích thước tập dữ liệu

```python
print("Shape =",df.shape)
```

Phân phối dữ liệu cột Tuổi

```python
plt.hist(df1.age, color="pink")
```

Phân phối dữ liệu cột BMI

```python
plt.hist(df1.bmi, color="brown")
```

Phân phối dữ liệu cột Giới tính

```python
plt.hist(df1.sex, color="blue")
```

Phân phối dữ liệu cột Hút thuốc

```python
plt.hist(df1.smoker, color="#102050")
```

Phân phối dữ liệu cột Khu vực

```python
plt.hist(df1.region, color="red")
```

Phân phối dữ liệu cột Trẻ em

```python
plt.hist(df1.children, color="green")
```

```python
plt.hist(df1.charges, color="gray")
```

Tương quan dữ liệu

```python
sns.heatmap(df.corr(), cmap='viridis',linewidths=2)
plt.savefig('Correlation_Heatmap.png')
plt.show()
```

Mô tả về các cột trong tập dữ liệu

```python
print(df.describe())
df.plot(kind='box')
```

Sử dụng cột Smoker làm bộ lọc => Mặc dù số người hút thuốc ít hơn, nhưng về mặt cơ bản, những người hút thuốc bị tính nhiều hơn

```python
sns.pairplot(df,hue='smoker')
plt.show()
```
