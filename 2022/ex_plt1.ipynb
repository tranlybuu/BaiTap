```python
import matplotlib.pyplot as plt
import numpy as np
```

![image](https://user-images.githubusercontent.com/70121634/157592565-47617044-711f-4dcd-94df-b95e9951306c.png)


```python
x_value = np.linspace(-6,5, num=1500)
y_value = x_value**3+x_value*3+7
y1_value = 3*(x_value**2)+3
y2_value = 6*x_value
```

```python
fig, (ax,ax1,ax2) = plt.subplots(1,3, figsize=(12,4))
ax.plot(x_value,y_value, 'tab:orange')
ax.set_title("Đồ thị y(x)")
ax.set_ylabel("Giá trị y")
ax1.plot(x_value,y1_value, 'tab:blue')
ax1.set_title("Đồ thị y'(x)")
ax1.set_xlabel("Giá trị x")
ax2.plot(x_value,y2_value, 'tab:gray')
ax2.set_title("Đồ thị y''(x)")
plt.show()
fig.savefig("ex_plt1.svg", format="svg")
```
