```python
import matplotlib.pyplot as plt
import numpy as np
```

### Khởi tạo mảng

```python
arr_name = ["A","B","C","D","E","F","G","H","L","M"]
arr_value = np.random.randint(1,10,10)
print(arr_value)
arr_value2 = np.random.randint(10,20,10)
print(arr_value2)
```

### LINE

```python
plt.plot(arr_value)
plt.show()
```

### BAR

```python
plt.bar(arr_name, arr_value)
plt.show()
```

```python
plt.barh(arr_name, arr_value, color="brown")
plt.show()
```

```python
index = np.arange(10)    
width = 0.3
plt.bar(index, arr_value, width, color='green',label='Cột giá trị random 1')
plt.bar(index+width, arr_value2, width, color='blue',label='Cột giá trị random 2')
plt.xticks(index+width/2, arr_name)     # Gắn nhãn cho trục x
plt.legend()    # Gắn nhãn chú thích
plt.show()
```

```python
""" Thêm đối số bottom và điều chỉnh lại khoảng cách các cột sẽ ra biểu đồ cột chồng """
index = np.arange(10)    
width = 0.3
plt.bar(index, arr_value, width, color='green',label='Cột giá trị random 1')
# plt.bar(index, arr_value2, width, color='blue',label='Cột giá trị random 2', bottom=arr_value)
plt.bar(index+width, arr_value2, width, color='blue',label='Cột giá trị random 2', bottom=arr_value)
plt.xticks(index+width/2, arr_name)     # Gắn nhãn cho trục x
plt.legend()    # Gắn nhãn chú thích
plt.show()
```

### PIE

```python
plt.pie(arr_value, labels=arr_name, counterclock=False, labeldistance=0.75)
plt.legend()
plt.axis('equal')
plt.show()
```

### Scatter

```python
plt.scatter(arr_value, arr_value2)
plt.show()
```

```python
""" Scatter 3D """
from mpl_toolkits import mplot3d
ax = plt.axes(projection="3d")
ax.scatter3D(arr_value, arr_value2)
plt.show()
```

### Đồ thị y=4x+5

```python
X_value = np.arange(-10,10,0.1)
Y_value = 4*X_value+5
plt.plot(X_value,Y_value)
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.title('Đồ thị y=4x+5')
plt.show()
```

### Đồ thị y=4xx+2x-1

```python
X_value = np.arange(-10,10,0.1)
Y_value = 4*(X_value**2+2*X_value-1)
plt.plot(X_value,Y_value)
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.title('Đồ thị y=4xx+2x-1')
plt.show()
```

### Đồ thị hàm Sin

```python
X_value = np.arange(0,4*np.pi,0.1)
Y_value = np.sin(X_value)
plt.plot(X_value,Y_value, "go")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.title('Đồ thị y=Sin(x)')
plt.show()
```
